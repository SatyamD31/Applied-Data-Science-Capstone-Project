## Download & Prepare Dataset

```python
import pandas as pd
```

```python
# fetching data from the given URL and storing in dataframe.
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
# original_url = 'https://en.wikipedia.org/w/index.php?title=List_of_postal_codes_of_Canada:_M&oldid=862527922'
df_list = pd.read_html(url)
df = df_list[0]
df.head()
```

```python
# eliminating 'Not assigned' boroughs
df = df[df.Borough != 'Not assigned'].reset_index(drop=True)
df.head()
```

```python
# group by Postal Code and combine neighbourhoods with same postal code
# df = df.groupby(['Postal Code', 'Borough'], as_index=False).agg(lambda x: ', '.join(x))
# df.head()
df['Neighborhood'] = df.groupby(['Postal Code','Borough'])['Neighborhood'].transform(lambda x: ', '.join(x))
df = df.drop_duplicates()
df.reset_index(drop = True, inplace=True)
df.head()
```

```python
# for neighborhoods="Not assigned", neighbourhood == borough
for index, row in df.iterrows():
    if row['Neighborhood'] == 'Not assigned':
        row['Neighborhood'] == row['Borough']
df.head()
```

```python
# converting the transformed dataset into the one as shown in the question

# create a new DataFrame
columns = ['Postal Code', 'Borough', 'Neighborhood']
temp_df = pd.DataFrame(columns=columns)

temp_list = ['M5G', 'M2H', 'M4B', 'M1J', 'M4G', 'M4M', 'M1R', 'M9V', 'M9L', 'M5V', 'M1B', 'M5A']

for pc in temp_list:
    temp_df = temp_df.append(df[df['Postal Code']==pc], ignore_index=True)

temp_df
```
