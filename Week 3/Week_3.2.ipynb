```python
import pandas as pd

url = 'https://cocl.us/Geospatial_data'
df_gps = pd.read_csv(url)
df_gps.head()
```

```python
df_gps.shape
```

```python
# git hub link to the raw format of csv file, which is the refined dataset, done in assignment 1
url = 'https://raw.githubusercontent.com/SidSat/IBM-DataScience_CapstoneProject/master/Dataset.csv'
df_data = pd.read_csv(url, index_col=0)
df_data.head()
```

```python
# merging dataset and their location coordinates
df_new = pd.merge(df_data, df_gps, on='Postal Code')
df_new.head()
```

```python
# converting the newly transformed dataset into the one as shown in the question

# create a new DataFrame
columns = ['Postal Code', 'Borough', 'Neighborhood', 'Latitude', 'Longitude']
temp_df = pd.DataFrame(columns=columns)

temp_list = ['M5G', 'M2H', 'M4B', 'M1J', 'M4G', 'M4M', 'M1R', 'M9V', 'M9L', 'M5V', 'M1B', 'M5A']

for pc in temp_list:
    temp_df = temp_df.append(df_new[df_new['Postal Code']==pc], ignore_index=True)

temp_df
```
